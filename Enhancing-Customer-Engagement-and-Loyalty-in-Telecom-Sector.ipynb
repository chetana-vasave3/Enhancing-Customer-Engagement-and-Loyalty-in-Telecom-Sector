when we are dealing with classification problems there are 2 senarios
here churners = yes or No
so first we have to identify yes:no ratio that is to find balanced or imbalnced data.
imbalanced data means the data is not 50:50.
when we have balanced data and we are creating our model on 50:50 ratio  that means our model is train on 50 records each on this classes so the model which is created here is definatley a efficient model.
bt if we have imbalnced data lets say 90:10 ratio obviously the model created will not be efficient coz 90 data has more info or more charactristics as compared to 10 so less info being transferred by this class. in this case overfitting will happened.



- impotant point

In a imbalaced dataset here i will be able to predict 90 record properly 10 records i will not be able to predict so 90% accuracy is not consider in this ..Accuracy is cursed in imbalaced dataset.

so there are some techniquesto deal with Imblnced dataset

- 1) upsampling: means creating senthetic records for low class.

e.g. suppose we have 5k non churners records and 2k churners records, now for churners we have 2k records we will senthetically create 5k records from this by using copy paste ...copy 2k records and create 3k copies out of it and just rename few features. so ultimately ur model has 7k records now ur model is being trained with 10 k records i.e. 5k for class 1 nad 5k for class 2.



- 2) down sampling: 
we have 2k records and we will arbitarily pick 2k records form upper class  now we build the model with this 4k records...


so it has been normaly seen that upsampling gives you better result than downsampling.coz in upsampling ypu have more data.

### 1.create a copy of base dataset for msnipulation and processing

```python
df_copy=df.copy()
```

### 2.Total charges should be numeric amount.lets convert it into numeric

### 3. As we can see there are 11 missing values lets check these records

```python
df_copy.loc[df_copy["TotalCharges"].isnull()==True]
```

### 4.Missinng value treatements

```python
## 11 records out of 7043 are NaN i.e 0.15% missing values
100*(11/7043)
```

since the % of records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing

```python
## Remove missing values
df_copy.dropna(how="any",inplace=True)
```

### 5.Divide customers into bins based on tenure e.g. for tenure <12 months: assign a tenure grp if 1-12,for tenure bw 1 to 2 to yrs, tenure grp of 13-24 ..so on...

```python
##get the max tenure
print(df_copy["tenure"].min())
print(df_copy["tenure"].max())
```

```python
##group the tenur in bins of 12 months
labels=["{0} -{1}".format(i,i+11) for i in range(1,72,12)]
df_copy["tenure_group"]=pd.cut(df_copy.tenure, range(1,80,12), right=False,labels=labels)
```

```python
def uniplot(df,col,title,hue=None):
    
    sns.set_style("whitegrid")
    sns.set_context("talk")
    plt.rcParams["axes.labelsize"]=20
    plt.rcParams["axes.titlesize"]=22
    plt.rcParams["axes.titlepad"]=30
    
    
    temp=pd.Series(data=hue)
    fig,ax=plt.subplots()
    width=len(df[col].unique())+ 7 + 4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation=45)
    plt.yscale("log")
    plt.title(title)
    ax= sns.countplot(data=df, x=col, order=df[col].value_counts().index, hue=hue, palette="bright")
    
    
    plt.show()
```

# Recommendations for Reducing Customer churn

- Targeted promotions for senior citizens: Offer customized promotions and plans specifically tailored to the needs and preferences of senior citizens. Highlight benefits such as senior discounts, simplified plans, and specialized customer support.

- Manage monthly charges and total charges: Provide tools or resources to help customers understand and manage their monthly and total charges. Offer budgeting advice, data usage monitoring, and alerts to prevent bill shock and ensure transparency in billing.

- Retention strategies for churned customers: offer incentives to win back customers who churned due to price concerns, service issues, or inadequate features.

- Enhance partner and dependents offerings:  provide bundled plans or family-oriented packages. Emphasize the advantages of shared plans, including discounted rates, shared data allowances, and family-friendly features.

- Optimize payment methods: Encourage customers to choose automatic payment methods such as bank transfer or credit card payments. Promote the benefits of convenience, timely payments, and potential discounts associated with automated payment options.

- Address tenure-based concerns: Analyze churn patterns based on tenure groups and identify specific challenges faced by customers in different stages of their relationship with the telecom service. Develop targeted strategies to address the unique needs and concerns of customers in each tenure group.

- Develop tailored marketing campaigns and offers for each segment to maximize engagement and satisfaction.

- Continuous customer satisfaction monitoring: Implement mechanisms to track and measure customer satisfaction continuously. Regularly collect feedback through surveys, focus groups, or sentiment analysis of customer interactions. Address any identified issues promptly to enhance overall satisfaction and reduce churn.

- Promote referral programs: Encourage satisfied customers to refer friends, family, and colleagues to the telecom service. Implement referral programs that reward both the existing customer and the new customer, such as discounts, account credits, or exclusive benefits.
